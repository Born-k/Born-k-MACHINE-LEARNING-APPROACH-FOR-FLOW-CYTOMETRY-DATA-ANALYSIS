```python
import numpy as np
import matplotlib.pyplot as plt
# from minisom import MiniSom
import seaborn as sns
from sklearn.cluster import Birch
from sklearn.cluster import AffinityPropagation
from sklearn.cluster import MeanShift
from sklearn.cluster import OPTICS
from sklearn_som.som import SOM
from matplotlib.colors import ListedColormap
from sklearn.metrics import silhouette_samples, silhouette_score,calinski_harabasz_score, davies_bouldin_score
```

```python
for i in range(4,10):
  som = SOM(m = i,n=1,dim=2)
  som.fit(temp)
  pred = som.predict(temp)
  silhouette_avg = silhouette_score(temp, pred)
  chs = calinski_harabasz_score(temp, pred)
  dbs = davies_bouldin_score(temp, pred)
  plt.figure(figsize = (6,6))
  plt.scatter(temp[:,0],
            temp[:,1],
            c = pred,
            cmap = "winter",
            marker = "*",
            s = 75)
  plt.show()
  print("silhouette Score: ", silhouette_avg)
  print("calinski harabasz Score: ", chs)
  print("davies bouldin Score: ", dbs)
```
